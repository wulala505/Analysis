```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

```python
df = pd.read_csv('product_price.csv',encoding='big5')
```

```python
df
```

```python
df = df.sort_values(by='price',ascending=False).reset_index(drop=True)
```

```python
df
```

```python
# 計算收入佔總收入百分比（累積）
p=[]
for i in range(len(df)):
    x = df['price'][i]/sum(df['price'])
    if(i!=0):
        p.append(x+p[i-1])
    else:
        p.append(x)
p
```

```python
fig,ax1 = plt.subplots()
ax1.bar(df.index,df['price'],align = 'center')
ax1.set_ylabel('收益')
ax1.set_xlabel('product')

# ax2共用ax1的x軸
ax2 = ax1.twinx()
ax2.plot(df.index,p,'r-')
ax2.set_ylabel('收入累積比例')
ax2.grid(False) #兩圖單位不同，格線先關掉

# 標註過80%的點
for i in range(len(p)):
    if(p[i]>0.8):
                     #round取第三位四捨五入小數點
        ax2.annotate(round(p[i],3),xy=(df.index[i],p[i]),xytext=(df.index[i],p[i]+0.08),  # xy =(箭頭指的點),xytext=(標註文字)
                                      arrowprops=dict(facecolor='black')) 
        print(df.index[i]) #第幾項超過80%收益
        break
                     
plt.xticks(df.index,df['product'])
plt.title('收益主次因素分析')

plt.show()                     
```

## 大同8月份363項有收益的商品，前70項就以佔了80%的收入(index從0開始)
